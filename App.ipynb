# Classify different means of transportation

With this app you can upload an image of a vehicle and the model will try to classify at it as one of the following categories:

Car, Bicycle, Motorcycle, Boat, Plane, Train, Helicopter, Scooter, Truck, Bus.

## Code

```
# Imports
#!pip install -Uqq fastai  # upgrade fastai on colab
from fastai.vision.all import *
from fastai.vision.widgets import *
from PIL import Image
from io import BytesIO
```

```
# Download learner from github releases, unzip and load the learner
download_url('https://github.com/T1mbo96/pds_a5_learner/archive/v1.0.zip', 'learn.pkl')
!unzip learn.pkl
learn_inf = load_learner('pds_a5_learner-1.0/export.pkl')
```

```
# Instantiate widgets
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description='Classify')

# On_click handler
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(512,512))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

## User Interface to use the app

```
VBox([widgets.Label('Upload your vehicle and classify it!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
